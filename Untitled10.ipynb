````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Activar estilo visual
sns.set_style("darkgrid")

# Definición de frecuencias y colores
frecuencias = [40, 80, 160]
colores = ['#1f77b4', '#ff7f0e', '#2ca02c']  # Azul, naranja, verde
frecuencia_fundamental = frecuencias[-1]
frecuencia_muestreo = frecuencia_fundamental * 15 + 1

# Tiempos de muestreo y observación
periodo_fundamental = 1 / frecuencia_fundamental
periodo_muestreo = 1 / frecuencia_muestreo
tiempo = np.arange(0, 40 * periodo_fundamental + periodo_muestreo, periodo_muestreo)

# Generar señales sinusoidales
componentes = [np.sin(2 * np.pi * f * tiempo) for f in frecuencias]
señal_compuesta = np.sum(componentes, axis=0)

# Preparar lista completa de señales y etiquetas
todas_las_senales = componentes + [señal_compuesta]
colores += ['crimson']  # Añadir color para la señal compuesta
titulos = [
    f'Senoidal {i+1}: {frecuencias[i]} Hz' for i in range(len(frecuencias))
] + ['Suma de las tres componentes']

# Crear figura con subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

for idx, (eje, señal, color, titulo) in enumerate(zip(axs.flat, todas_las_senales, colores, titulos)):
    eje.plot(tiempo, señal, color=color, linewidth=1.5)
    eje.set_title(titulo, fontsize=11)
    eje.set_xlabel('Tiempo (s)')
    eje.set_ylabel('Amplitud')
    eje.grid(True)

fig.tight_layout()
plt.show()
````

Calcule la transformada de Fourier de la señal y grafique el espectro de frecuencia.
Identifique en el espectro las frecuencias que componen la señal (trabaje con 10 periodos
de la señal). (10%)

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo visual
sns.set_style("whitegrid")

# Transformada de Fourier
X = np.fft.fft(señal_compuesta)
N = len(señal_compuesta)
frec_muestreo = 1 / periodo_muestreo
f = np.fft.fftfreq(N, d=periodo_muestreo)
magnitudes = np.abs(X)

# Solo la mitad positiva del espectro
frecuencias_positivas = f[:N // 2]
magnitudes_positivas = magnitudes[:N // 2]

# Crear figura
plt.figure(figsize=(12, 6))
plt.plot(frecuencias_positivas, magnitudes_positivas, color='#1f77b4', linewidth=1.8)
plt.title("Espectro de Frecuencia de la Señal Compuesta", fontsize=13)
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.xlim(0, 200)

# Marcar frecuencias componentes
frecuencias_a_resaltar = [40, 80, 160]
for freq in frecuencias_a_resaltar:
    plt.axvline(x=freq, color='crimson', linestyle='--', linewidth=1)
    plt.text(freq, max(magnitudes_positivas) * 0.05, f'{freq} Hz',
             color='crimson', ha='center', fontsize=10)

plt.tight_layout()
plt.show()
````

La transformada inversa de Fourier es una herramienta matemática clave que permite recuperar una señal en el dominio del tiempo a partir de su representación en el dominio de la frecuencia. Actúa como el proceso complementario de la transformada de Fourier, la cual descompone una señal temporal en sus diferentes componentes de frecuencia.

Esta operación es ampliamente utilizada en áreas como la ingeniería, la física y el procesamiento de señales, ya que posibilita no solo el análisis, sino también la reconstrucción de señales modificadas o procesadas en el dominio de la frecuencia. En particular, es muy útil cuando se aplican técnicas de filtrado: una señal puede ser transformada al dominio de la frecuencia para atenuar o eliminar componentes no deseadas, como el ruido, y luego, mediante la transformada inversa, se obtiene nuevamente la señal modificada en el dominio del tiempo, lista para ser analizada, visualizada o utilizada.

````python
# Inversa de Fourier
x_inv = np.fft.ifft(X)

# Graficar la señal
plt.figure(figsize=(10, 6))
plt.plot(tiempo, np.real(x_inv))
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.title("Señal reconstruida mediante IFFT")
plt.show()
````

El archivo adjunto (senecg.mat) contiene una señal de ECG adquirida a una frecuencia de muestreo de 250 Hz. Realice un análisis en frecuencia de la señal usando periodogramas de Welch y determine si es necesario eliminar ruido, en tal caso, elimínelo de la forma que se mostró antes y compruebe que lo haya realizado, calcule la transformada inversa de Fourier de la señal filtrada y compruebe con el periodograma. (Usar una ventana Hanning, y justificar la elección del solapamiento y el ancho de la ventaja) (25%)

````python
# Estilo visual
sns.set_style("whitegrid")

# Cargar archivo .mat y extraer la señal ECG
file = "senecg.mat"
data = scipy.io.loadmat(file)
ecg_signal = data['ECG'].flatten()

# Frecuencia de muestreo (Hz)
fs = 250

# Calcular el periodograma de Welch
frecuencias, Pxx = welch(ecg_signal, fs, window='hann', nperseg=256, noverlap=128)

plt.figure(figsize=(12, 6))
plt.semilogy(frecuencias, Pxx, color='#1f77b4', linewidth=2)

# gráfico
plt.title('Estimación espectral con el método de Welch', fontsize=14)
plt.xlabel('Frecuencia (Hz)', fontsize=12)
plt.ylabel('Densidad espectral de potencia (V²/Hz)', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xlim(0, fs / 2)
plt.tight_layout()
plt.show()
````

Al observar el espectro generado mediante el método de Welch, se evidencia que la mayor parte de la energía de la señal se concentra en las frecuencias bajas, especialmente por debajo de los 40 Hz. Este comportamiento coincide con lo reportado en estudios sobre señales ECG, donde se ha establecido que las componentes de interés clínico suelen encontrarse entre 0.5 Hz y 40 Hz. Por ello, se adoptó un límite inferior de 0.5 Hz para el filtrado pasa banda, asegurando la retención de información fisiológicamente relevante [1].

Por otro lado, se observó un pico marcado en 60 Hz, típico de la interferencia de la red eléctrica, un problema común en ambientes hospitalarios o domésticos con equipos electrónicos cercanos. Este tipo de ruido puede afectar la precisión del análisis, especialmente si se confunde con actividad eléctrica del corazón. Dado que esta frecuencia está fuera del rango de interés del ECG, se optó por un filtro de paso bajo que reduce su influencia sin alterar las componentes útiles de la señal [2].

Este enfoque de limpieza espectral permite trabajar con una señal más clara, minimizando artefactos y mejorando la calidad de los resultados. Filtrar adecuadamente garantiza que los análisis se enfoquen en las bandas que verdaderamente contienen información diagnóstica, lo que es esencial para detectar arritmias, variabilidad de la frecuencia cardíaca u otras anomalías clínicas [3].

````python
T_muestreo = 1 / fs
t = np.arange(0, len(ecg_signal) / fs, T_muestreo)

# Transformada de Fourier
F1 = np.fft.fft(ecg_signal)
frecuencias_fft = np.fft.fftfreq(len(F1), d=T_muestreo)

# Aplicación de filtro pasa banda (0.5 Hz a 40 Hz)
F1_filtrada = np.zeros_like(F1, dtype=complex)
for i in range(len(frecuencias_fft)):
    if 0.5 <= abs(frecuencias_fft[i]) <= 40:
        F1_filtrada[i] = F1[i]

# Transformada inversa para obtener señal filtrada en el tiempo
ecg_filtrado = np.fft.ifft(F1_filtrada)

# Preparar para graficar señales original y filtrada
señales = [ecg_signal, np.real(ecg_filtrado)]
titulos = [
    'Señal ECG Original',
    'Señal ECG Filtrada (0.5–40 Hz, ruido de 60 Hz eliminado)'
]
colores = ['#2ca02c', '#d62728']

# Graficar ambas señales
plt.figure(figsize=(14, 8))

for i, señal in enumerate(señales):
    plt.subplot(2, 1, i + 1)
    plt.plot(t, señal, color=colores[i], linewidth=1.2)
    plt.title(titulos[i], fontsize=14)
    plt.xlabel('Tiempo (s)', fontsize=12)
    plt.ylabel('Amplitud', fontsize=12)
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
````

````python
frecuencias_welch, Pxx_welch = welch(ecg_filtrado, fs, window='hann', nperseg=256, noverlap=128)
plt.figure(figsize=(12, 6))
plt.semilogy(frecuencias_welch, Pxx_welch)
plt.title('Periodograma de Welch de la Señal Filtrada')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Densidad de Potencia (V^2/Hz)')
plt.grid()
plt.tight_layout()
plt.show()
````
